## Documentation for Converting the Naturalizations Column to int64 Type

### Objective:
To convert the data in the **`Naturalizations`** column from **`object`** (string) type to **`int64`** type to enable statistical analysis and visualization.



**Inspect Unique Values:**
   - Use `data['Naturalizations'].unique()` to check the unique values in the **`Naturalizations`**
   
    column to understand the format of the data (such as any spaces or special characters).


````python
print(data['Naturalizations'].unique())
````


**Clean the Data:**
   - Remove any extra spaces, commas, or unnecessary symbols using the **`str.replace()`** method.
   - Use **`str.strip()`** to remove any leading or trailing spaces.
   

````python
data['Naturalizations'] = data['Naturalizations'].str.replace(' ', '')
data['Naturalizations'] = data['Naturalizations'].str.replace(',', '')
data['Naturalizations'] = data['Naturalizations'].str.strip()
````


**Convert the Column to Numeric Type:**
   - Use **`pd.to_numeric()`** to convert the **`Naturalizations`** column to **`int64 | float64`**  type. Use `errors='coerce'` to convert any non-convertible values to **`NaN`**.


````python
data['Naturalizations'] = pd.to_numeric(data['Naturalizations'], errors='coerce')
````


**Verify the Conversion:**
   - Use `data.dtypes` to ensure that the **`Naturalizations`** column has been successfully converted to **`int64 | float64`** type.


````python
print(data.dtypes)
````


### Outcome:
After completing these steps, the **`Naturalizations`** column is ready for use in statistical analyses and visualizations, facilitating a better understanding of the data and enabling data-driven decision-making.

# Scaling


❐ Scaling procedure for data to improve the performance of statistical models and machine learning

by standardizing the range of values, enhancing accuracy and efficiency in analysis






### **Method Used:**
- **Robust Scaling**  
  This method scales the data by removing the median and then dividing by the interquartile range (IQR). It is **less sensitive to outliers** compared to other scaling methods like Z-Score Scaling, making it ideal for datasets with extreme values.

---



````python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
data_scaled = scaler.fit_transform(data)
data = pd.DataFrame(data_scaled, columns=data.columns)
````
